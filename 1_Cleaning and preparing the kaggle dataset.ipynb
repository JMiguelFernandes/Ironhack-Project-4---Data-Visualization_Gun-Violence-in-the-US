## Expanding columns

Some of the columns contained multiple values related to the several people involved in an incident or the several guns involved. 

I want to use some of that data in Tableau, so I wrote a function to do the following: 
1. Expand the column into a dataframe
2. Transform the dataframe from a wide format to a long one (so that it's nice and tidy for Tableau).

```python
def expand_column(column, index, names):
    column_expanded = (column
                       .fillna("not available")
                       .str.replace("::", "||")
                       .str.split("\|\|"))
    
    column_expanded = pd.DataFrame(column_expanded.to_list())
    
    column_expanded = column_expanded.assign(incident_id = index)
    
    column_expanded = pd.melt(column_expanded, 
                              id_vars="incident_id", 
                              value_vars=np.arange(1, column_expanded.shape[1], 2),
                              var_name = names[0],
                              value_name = names[1]).dropna()
    return(column_expanded)
```

## Creating a table with participant data

I created a long-format table with info regarding participant gender, age, type and status and saved it as a csv file.

```python
participant_gender_tidy = expand_column(gv["participant_gender"], gv["incident_id"], ("participant_number", "participant_gender"))
```

```python
participant_age_tidy = expand_column(gv["participant_age"], 
                                     gv["incident_id"], 
                                     ("participant_number", "participant_age"))
```

```python
participant_type_tidy = expand_column(gv["participant_type"], 
                                      gv["incident_id"], 
                                      ("participant_number", "participant_type"))
```

```python
participant_status_tidy = expand_column(gv["participant_status"], 
                                        gv["incident_id"], 
                                        ("participant_number", "participant_status"))
```

```python
participant_type_and_gender = participant_type_tidy.merge(participant_gender_tidy, 
                                                          on = ["incident_id", "participant_number"], 
                                                          how= "left")
```

```python
participant_type_and_gender
```

```python
participant_type_gender_and_age = participant_type_and_gender.merge(participant_age_tidy, 
                                                          on = ["incident_id", "participant_number"], 
                                                          how= "left")
```

```python
participant_type_gender_and_age
```

```python
participant_type_gender_age_and_status = participant_type_gender_and_age.merge(participant_status_tidy, 
                                                          on = ["incident_id", "participant_number"], 
                                                          how= "left")
```

```python
participant_type_gender_age_and_status
```

```python
participant_type_gender_age_and_status.to_csv("data_clean/participant_tgas.csv", index = False)
```

## Creating a table with participant data

I did the same for the gun info (whether they were stolen and what type of gun).

```python
gun_stolen = expand_column(gv["gun_stolen"], gv["incident_id"], ("gun_number", "gun_stolen"))
gun_type = expand_column(gv["gun_type"], gv["incident_id"], ("gun_number", "gun_type"))
```

```python
gun_stolen_and_type = gun_type.merge(gun_stolen, 
                                     on = ["incident_id", "gun_number"], 
                                     how= "left")
```

```python
gun_stolen_and_type
```

```python
gun_stolen_and_type.to_csv("data_clean/gun_info.csv", index = False)
```
